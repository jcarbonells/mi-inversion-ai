```python
# ============================================
# 09_orchestrator.ipynb - Automatizaci√≥n y Agente GenAI (MEJORADO + CORREGIDO)
# ============================================

!pip -q install pandas google-generativeai

import os
import pandas as pd
from datetime import datetime
from google.colab import drive
import google.generativeai as genai
import json
import glob  # <-- A√±adido para validar reporter

# --- Montar Drive ---
drive.mount('/content/drive', force_remount=False)

# --- Configurar GenAI ---
GEMINI_API_KEY = "AIzaSyCjID5QZSe0xDvGaq7mTHcHOctWBaxaAn8"
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel('models/gemini-2.0-flash')

BASE = "/content/drive/MyDrive/investment_ai"
NOTEBOOKS_DIR = f"{BASE}/notebooks"
REPORTS_DIR = f"{BASE}/reports"

# --- Lista de notebooks en orden CORRECTO ---
NOTEBOOKS = [
    "00_liquidity_agent.ipynb",      # ‚úÖ NUEVO: Primero (datos macro independientes)
    "01_data_prep.ipynb",
    "02_portfolio_exposure.ipynb",
    "06_portfolio_reconstructor.ipynb",
    "07_asset_metrics.ipynb",
    "11_market_analyst.ipynb",
    "12_sectorial_strength.ipynb",
    "03_fx_agent.ipynb",
    "04_quant_signals.ipynb",
    "05_risk_manager.ipynb",
    "07_reporter_advanced.ipynb"
]

# --- Archivos cr√≠ticos que deben existir ANTES de ciertos agentes ---
DEPENDENCIAS = {
    "03_fx_agent.ipynb": ["market_regime_latest.json", "portfolio_daily_value.csv"],
    "04_quant_signals.ipynb": ["market_regime_latest.json", "sector_strength_latest.json", "portfolio_enriched_final.csv"],
    "05_risk_manager.ipynb": ["portfolio_daily_value.csv", "portfolio_enriched_final.csv"]
}

# --- Funci√≥n para verificar dependencias ---
def verificar_dependencias(notebook_name):
    if notebook_name not in DEPENDENCIAS:
        return True
    for archivo in DEPENDENCIAS[notebook_name]:
        ruta = os.path.join(REPORTS_DIR, archivo)
        if not os.path.exists(ruta):
            print(f"  ‚ö†Ô∏è Dependencia faltante: {archivo} (requerido por {notebook_name})")
            return False
    return True

# --- Funci√≥n para validar salidas ---
def validar_salidas(notebook_name):
    """Valida que el notebook gener√≥ sus archivos de salida esperados."""
    contratos = {
        "00_liquidity_agent.ipynb": [
            "reports/liquidity_regime_latest.json",
            "reports/liquidity_dashboard_latest.md",
            "reports/liquidity_dashboard_latest.html"
        ],
        "01_data_prep.ipynb": ["data/raw/my_portfolio_prices.parquet"],
        "02_portfolio_exposure.ipynb": ["reports/portfolio_enriched_final.csv"],
        "06_portfolio_reconstructor.ipynb": ["reports/portfolio_daily_value.csv"],
        "07_asset_metrics.ipynb": ["reports/asset_metrics.csv"],
        "11_market_analyst.ipynb": ["reports/market_regime_latest.json"],
        "12_sectorial_strength.ipynb": ["reports/sector_strength_latest.json"],
        "03_fx_agent.ipynb": ["reports/fx_hedge_latest.json"],
        "04_quant_signals.ipynb": ["reports/quant_signals_latest.json"],
        "05_risk_manager.ipynb": ["reports/risk_dashboard_latest.json"],
    }

    # ‚úÖ Manejo especial para el reporter
    if notebook_name == "07_reporter_advanced.ipynb":
        report_pattern = os.path.join(REPORTS_DIR, "report_final_*.md")
        if not glob.glob(report_pattern):
            print(f"  ‚ö†Ô∏è Archivo de salida faltante: {report_pattern}")
            return False
        return True

    if notebook_name not in contratos:
        return True

    for archivo in contratos[notebook_name]:
        ruta = os.path.join(BASE, archivo)
        if not os.path.exists(ruta):
            print(f"  ‚ö†Ô∏è Archivo de salida faltante: {ruta}")
            return False
    return True

# --- Funci√≥n para ejecutar un notebook con %run ---
def run_notebook(notebook_name):
    if not verificar_dependencias(notebook_name):
        return False

    path = f"{NOTEBOOKS_DIR}/{notebook_name}"
    if not os.path.exists(path):
        print(f"  ‚ö†Ô∏è {notebook_name} no encontrado en {NOTEBOOKS_DIR}")
        return False

    try:
        print(f"  ‚úÖ Ejecutando {notebook_name}...")
        %run $path

        if not validar_salidas(notebook_name):
            print(f"    ‚ùå {notebook_name} no gener√≥ los archivos esperados.")
            return False

        return True
    except Exception as e:
        print(f"    ‚ùå Error: {str(e)[:200]}...")
        return False

def generate_ai_dashboard_summary():
    """
    Genera un an√°lisis exhaustivo de la cartera usando GenAI,
    con sem√°foro, recomendaci√≥n y justificaci√≥n por indicador.
    """
    print("üß† Generando an√°lisis exhaustivo con GenAI...")

    # --- Cargar datos con manejo robusto ---
    pf_path = f"{REPORTS_DIR}/portfolio_enriched_final.csv"
    pf = pd.read_csv(pf_path) if os.path.exists(pf_path) else pd.DataFrame()

    risk_path = f"{REPORTS_DIR}/risk_dashboard.csv"
    risk = pd.read_csv(risk_path) if os.path.exists(risk_path) else pd.DataFrame()

    quant_path = f"{REPORTS_DIR}/quant_signals.csv"
    quant = pd.read_csv(quant_path) if os.path.exists(quant_path) else pd.DataFrame()

    fx_path = f"{REPORTS_DIR}/fx_hedge_signal.csv"
    fx = pd.read_csv(fx_path) if os.path.exists(fx_path) else pd.DataFrame()

    metrics_path = f"{REPORTS_DIR}/asset_metrics.csv"
    metrics = pd.read_csv(metrics_path) if os.path.exists(metrics_path) else pd.DataFrame()

    sector_path = f"{REPORTS_DIR}/sector_scores_latest.csv"
    sectors = pd.read_csv(sector_path) if os.path.exists(sector_path) else pd.DataFrame()

    breaches_asset_path = f"{REPORTS_DIR}/breaches_asset.csv"
    breaches_asset = pd.DataFrame()
    if os.path.exists(breaches_asset_path):
        try:
            breaches_asset = pd.read_csv(breaches_asset_path, encoding='utf-8-sig')
            breaches_asset.columns = breaches_asset.columns.str.strip()
        except Exception as e:
            print(f"‚ö†Ô∏è Error al leer breaches_asset.csv: {e}")

    # --- Extraer m√©tricas clave ---
    total = pf["importe_actual_eur"].sum() if not pf.empty else 0
    drawdown = "N/A"
    alpha = "N/A"
    retorno_cartera = "N/A"
    sp500_retorno = "N/A"

    if not risk.empty and 'm√©trica' in risk.columns:
        def get_metric_value(metric_name):
            row = risk[risk['m√©trica'] == metric_name]
            return row['valor'].values[0] if not row.empty else "N/A"
        drawdown = get_metric_value('Drawdown actual')
        alpha = get_metric_value('Alpha vs S&P 500')
        retorno_cartera = get_metric_value('Retorno anualizado cartera')
        sp500_retorno = get_metric_value('Retorno anualizado S&P 500')

    activos_concentrados = "N/A"
    regiones_concentradas = "N/A"
    if not breaches_asset.empty and 'm√©trica' in breaches_asset.columns:
        activos_row = breaches_asset[breaches_asset['m√©trica'] == 'Activos > 5%']
        regiones_row = breaches_asset[breaches_asset['m√©trica'] == 'Regiones > 30%']
        if not activos_row.empty:
            activos_concentrados = activos_row['valor'].values[0]
        if not regiones_row.empty:
            regiones_concentradas = regiones_row['valor'].values[0]

    usd_no_hedged = "N/A"
    if not breaches_asset.empty and 'USD no hedged (%)' in breaches_asset['m√©trica'].values:
        usd_row = breaches_asset[breaches_asset['m√©trica'] == 'USD no hedged (%)']
        usd_no_hedged = usd_row['valor'].values[0] if not usd_row.empty else "N/A"

    # --- Construir prompt para GenAI ---
    prompt = f"""
Eres un agente de inversi√≥n profesional con acceso a un sistema multiagente avanzado.
Tu tarea es generar un **an√°lisis exhaustivo de la cartera al {datetime.today().strftime('%Y-%m-%d')}** en formato de tabla con las siguientes columnas:

- **Categor√≠a**: riesgo, rendimiento, concentraci√≥n, FX, sectorial, se√±ales, etc.
- **Indicador**: nombre del indicador clave.
- **Valor**: valor num√©rico o categ√≥rico.
- **Sem√°foro**: usa ‚úÖ (verde), ‚ö†Ô∏è (√°mbar) o ‚ùå (rojo).
- **¬øVamos bien?**: "S√≠", "No" o "Parcialmente".
- **Recomendaci√≥n**: acci√≥n concreta (ej. "Rebalancear", "Aumentar XLK", "Cubrir 10% USD").
- **Justificaci√≥n**: explica por qu√©, citando el indicador y su valor.

**Reglas estrictas**:
- Usa **solo los datos proporcionados**.
- S√© **conciso, preciso y accionable**.
- No inventes datos.
- El tono debe ser profesional pero claro para un inversor no t√©cnico.

**Datos de la cartera**:
- Valor total: {total:,.0f} ‚Ç¨
- Drawdown actual: {drawdown}
- Retorno anualizado cartera: {retorno_cartera}
- Retorno anualizado S&P 500: {sp500_retorno}
- Alpha vs S&P 500: {alpha}
- Activos >5% del portafolio: {activos_concentrados}
- Regiones >30%: {regiones_concentradas}
- USD no hedged: {usd_no_hedged}%

**Se√±ales sectoriales (Symbol, Score, RSC, Signal)**:
{sectors.to_string(index=False) if not sectors.empty else "No disponible"}

**Se√±ales cuantitativas (ejemplos)**:
{quant.head(5).to_string(index=False) if not quant.empty else "No disponible"}

**Brechas de riesgo**:
- Drawdown ‚â•9%: {'S√≠' if 'True' in str(drawdown) else 'No'}
- Alpha objetivo (+5%): {'Cumplido' if alpha != 'N/A' and float(str(alpha).replace('%','').replace(',','.')) >= 5 else 'Incumplido'}

Genera la tabla en **formato Markdown**. No a√±adas introducci√≥n ni conclusi√≥n, solo la tabla.
"""

    try:
        response = model.generate_content(prompt)
        analysis_md = response.text

        output_path = f"{REPORTS_DIR}/ai_dashboard_summary_{datetime.today().strftime('%Y%m%d')}.md"
        with open(output_path, "w", encoding="utf-8") as f:
            f.write("# üìä An√°lisis Exhaustivo de Cartera (GenAI)\n\n")
            f.write(f"Fecha: {datetime.today().strftime('%Y-%m-%d')}\n\n")
            f.write(analysis_md)

        print(f"‚úÖ An√°lisis guardado en: {output_path}")
        return analysis_md

    except Exception as e:
        error_msg = f"‚ö†Ô∏è Error al generar an√°lisis con GenAI: {e}"
        print(error_msg)
        return error_msg

# --- Funci√≥n para ejecutar la cadena semanal ---
def run_weekly_pipeline():
    print("üîÑ Iniciando cadena semanal de agentes...")
    log = []
    start_time = datetime.now()

    for nb in NOTEBOOKS:
        success = run_notebook(nb)
        log.append(f"{nb}: {'OK' if success else 'FALL√ì'}")
        if not success:
            print("  ‚ö†Ô∏è Cadena interrumpida por error.")
            break

    end_time = datetime.now()
    duration = (end_time - start_time).total_seconds()
    log_path = f"{REPORTS_DIR}/orchestrator_log_{start_time.strftime('%Y%m%d')}.txt"
    with open(log_path, "w") as f:
        f.write(f"Cadena ejecutada: {start_time}\n")
        f.write(f"Duraci√≥n: {duration:.0f} segundos\n")
        f.write("\n".join(log))

    print(f"‚úÖ Cadena completada en {duration:.0f} segundos. Log guardado en {log_path}")

    if all("OK" in entry for entry in log):
        print("üß† Generando an√°lisis exhaustivo con GenAI...")
        generate_ai_dashboard_summary()
    else:
        print("‚ö†Ô∏è No se gener√≥ el an√°lisis de GenAI debido a errores en la cadena.")

# --- Depuraci√≥n: verificar columnas del risk_dashboard ---
risk_debug_path = f"{REPORTS_DIR}/risk_dashboard.csv"
if os.path.exists(risk_debug_path):
    risk_debug = pd.read_csv(risk_debug_path)
    print(f"üîç Columnas en risk_dashboard.csv: {list(risk_debug.columns)}")

# --- Funci√≥n para responder preguntas ---
def answer_question(user_question):
    try:
        # --- Cargar datos existentes con manejo de errores ---
        pf_path = f"{REPORTS_DIR}/portfolio_enriched_final.csv"
        pf = pd.read_csv(pf_path) if os.path.exists(pf_path) else pd.DataFrame()

        risk_path = f"{REPORTS_DIR}/risk_dashboard.csv"
        risk = pd.read_csv(risk_path) if os.path.exists(risk_path) else pd.DataFrame()

        # --- Depuraci√≥n: mostrar columnas reales ---
        if not risk.empty:
            print(f"üîç Columnas reales en risk_dashboard: {list(risk.columns)}")

        quant_path = f"{REPORTS_DIR}/quant_signals.csv"
        quant = pd.read_csv(quant_path) if os.path.exists(quant_path) else pd.DataFrame()

        fx_path = f"{REPORTS_DIR}/fx_hedge_signal.csv"
        fx = pd.read_csv(fx_path) if os.path.exists(fx_path) else pd.DataFrame()

        metrics_path = f"{REPORTS_DIR}/asset_metrics.csv"
        metrics = pd.read_csv(metrics_path) if os.path.exists(metrics_path) else pd.DataFrame()

        # --- Encontrar la columna correcta para m√©trica ---
        metric_col = None
        value_col = None

        if not risk.empty:
            # Buscar columna que contenga 'm√©trica', 'metrica', o similar
            for col in risk.columns:
                if 'm√©trica' in col.lower() or 'metrica' in col.lower() or 'metric' in col.lower():
                    metric_col = col
                    break

            # Buscar columna que contenga 'valor'
            for col in risk.columns:
                if 'valor' in col.lower() or 'value' in col.lower():
                    value_col = col
                    break

        # --- Cargar contexto MACRO ---
        macro_str = "No disponible"
        macro_path = f"{REPORTS_DIR}/market_regime_latest.json"
        if os.path.exists(macro_path):
            with open(macro_path) as f:
                macro = json.load(f)
            macro_str = f"{macro['regimen']} (Score: {macro['score']:.2f}, Percentil: {macro['percentil']:.0f}%)"

        # --- Cargar fuerza SECTORIAL ---
        sectors_str = "No disponible"
        sectors_path = f"{REPORTS_DIR}/sector_strength_latest.json"
        if os.path.exists(sectors_path):
            with open(sectors_path) as f:
                sectors_data = json.load(f)
            if sectors_data.get("top_oportunidades"):
                sectors_str = "; ".join(sectors_data["top_oportunidades"])

        # --- Extraer KPIs de riesgo ---
        drawdown = "N/A"
        alpha = "N/A"

        if not risk.empty and metric_col and value_col:
            # Buscar Drawdown actual
            dd_rows = risk[risk[metric_col].astype(str).str.contains('Drawdown', case=False, na=False)]
            if not dd_rows.empty:
                drawdown = dd_rows[value_col].iloc[0]

            # Buscar Alpha
            alpha_rows = risk[risk[metric_col].astype(str).str.contains('Alpha', case=False, na=False)]
            if not alpha_rows.empty:
                alpha = alpha_rows[value_col].iloc[0]

        # --- Resto del c√≥digo igual ---
        total = pf["importe_actual_eur"].sum() if not pf.empty else 0
        quant_summary = "Ninguna" if quant.empty else ", ".join([f"{row['Activo']}: {row['Se√±al']}" for _, row in quant.iterrows()])
        fx_summary = "Ninguna" if fx.empty else ", ".join([f"{row['divisa']}: {row['cobertura_recomendada']} ({row['%_cobertura']}%)" for _, row in fx.iterrows()])
        acn_alpha = "N/A"
        if not metrics.empty and "ACN" in metrics["Activo"].values:
            acn_row = metrics[metrics["Activo"] == "ACN"].iloc[0]
            acn_alpha = f"{acn_row['Retorno total mejorado']:.1%}" if pd.notna(acn_row.get('Retorno total mejorado')) else "N/A"

        context = f"""
        Eres un asistente de inversi√≥n personal. Responde usando SOLO estos datos:

        **Contexto Macro Actual:**
        - R√©gimen: {macro_str}

        **Fuerza Sectorial (Top oportunidades):**
        - {sectors_str}

        **Cartera Global:**
        - Valor total: {total:,.0f} ‚Ç¨
        - Drawdown: {drawdown}
        - Alpha vs S&P 500: {alpha}

        **Se√±ales Cuantitativas:**
        - {quant_summary}

        **Se√±ales FX:**
        - {fx_summary}

        **Insights por Activo:**
        - Alpha de ACN (incluyendo regalos): {acn_alpha}

        Pregunta del usuario: {user_question}
        """

        response = model.generate_content(context)
        return response.text

    except Exception as e:
        return f"‚ö†Ô∏è Error al cargar datos: {e}"

# --- Interfaz principal ---
print("üß† Orchestrator + Agente GenAI Activo")
print("Comandos:")
print("- 'semanal': ejecuta la cadena completa de agentes")
print("- 'pregunta [tu pregunta]': interact√∫a con GenAI")
print("- 'salir': termina")

while True:
    user_input = input("\n> ").strip()

    if user_input == "salir":
        break
    elif user_input == "semanal":
        run_weekly_pipeline()
    elif user_input.startswith("pregunta "):
        question = user_input[9:]
        answer = answer_question(question)
        print(f"\nü§ñ {answer}")
    else:
        print("‚ùì Comando no reconocido. Usa 'semanal', 'pregunta ...' o 'salir'.")

# üîî RECOMENDACI√ìN FINAL:
# Aseg√∫rate de que todos tus agentes guarden archivos CSV con:
# df.to_csv("archivo.csv", encoding='utf-8-sig', index=False)
# Esto evita errores de BOM y columnas mal le√≠das.
```

```python
# ============================================
# 09_orchestrator.ipynb - Automatizaci√≥n y Agente GenAI (MEJORADO + PDF MAESTRO)
# ============================================

!pip -q install pandas google-generativeai markdown weasyprint

import os
import pandas as pd
from datetime import datetime
from google.colab import drive
import google.generativeai as genai
import json
import glob

# --- Montar Drive ---
drive.mount('/content/drive', force_remount=False)

# --- Configurar GenAI ---
GEMINI_API_KEY = "AIzaSyCjID5QZSe0xDvGaq7mTHcHOctWBaxaAn8"
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel('models/gemini-2.0-flash')

BASE = "/content/drive/MyDrive/investment_ai"
NOTEBOOKS_DIR = f"{BASE}/notebooks"
REPORTS_DIR = f"{BASE}/reports"

# --- Lista de notebooks en orden CORRECTO ---
NOTEBOOKS = [
    "00_liquidity_agent.ipynb",
    "01_data_prep.ipynb",
    "02_portfolio_exposure.ipynb",
    "06_portfolio_reconstructor.ipynb",
    "07_asset_metrics.ipynb",
    "11_market_analyst.ipynb",
    "12_sectorial_strength.ipynb",
    "03_fx_agent.ipynb",
    "04_quant_signals.ipynb",
    "05_risk_manager.ipynb",
    "07_reporter_advanced.ipynb"
]

# --- Archivos cr√≠ticos que deben existir ANTES de ciertos agentes ---
DEPENDENCIAS = {
    "03_fx_agent.ipynb": ["market_regime_latest.json", "portfolio_daily_value.csv"],
    "04_quant_signals.ipynb": ["market_regime_latest.json", "sector_strength_latest.json", "portfolio_enriched_final.csv"],
    "05_risk_manager.ipynb": ["portfolio_daily_value.csv", "portfolio_enriched_final.csv"]
}

# --- Funci√≥n para verificar dependencias ---
def verificar_dependencias(notebook_name):
    if notebook_name not in DEPENDENCIAS:
        return True
    for archivo in DEPENDENCIAS[notebook_name]:
        ruta = os.path.join(REPORTS_DIR, archivo)
        if not os.path.exists(ruta):
            print(f"  ‚ö†Ô∏è Dependencia faltante: {archivo} (requerido por {notebook_name})")
            return False
    return True

# --- Funci√≥n para validar salidas ---
def validar_salidas(notebook_name):
    """Valida que el notebook gener√≥ sus archivos de salida esperados."""
    contratos = {
        "00_liquidity_agent.ipynb": [
            "reports/liquidity_regime_latest.json",
            "reports/liquidity_dashboard_latest.md",
            "reports/liquidity_dashboard_latest.html"
        ],
        "01_data_prep.ipynb": ["data/raw/my_portfolio_prices.parquet"],
        "02_portfolio_exposure.ipynb": ["reports/portfolio_enriched_final.csv"],
        "06_portfolio_reconstructor.ipynb": ["reports/portfolio_daily_value.csv"],
        "07_asset_metrics.ipynb": ["reports/asset_metrics.csv"],
        "11_market_analyst.ipynb": ["reports/market_regime_latest.json"],
        "12_sectorial_strength.ipynb": ["reports/sector_strength_latest.json"],
        "03_fx_agent.ipynb": ["reports/fx_hedge_latest.json"],
        "04_quant_signals.ipynb": ["reports/quant_signals_latest.json"],
        "05_risk_manager.ipynb": ["reports/risk_dashboard_latest.json"],
    }

    if notebook_name == "07_reporter_advanced.ipynb":
        report_pattern = os.path.join(REPORTS_DIR, "report_final_*.md")
        if not glob.glob(report_pattern):
            print(f"  ‚ö†Ô∏è Archivo de salida faltante: {report_pattern}")
            return False
        return True

    if notebook_name not in contratos:
        return True

    for archivo in contratos[notebook_name]:
        ruta = os.path.join(BASE, archivo)
        if not os.path.exists(ruta):
            print(f"  ‚ö†Ô∏è Archivo de salida faltante: {ruta}")
            return False
    return True

# --- Funci√≥n para ejecutar un notebook con %run ---
def run_notebook(notebook_name):
    if not verificar_dependencias(notebook_name):
        return False

    path = f"{NOTEBOOKS_DIR}/{notebook_name}"
    if not os.path.exists(path):
        print(f"  ‚ö†Ô∏è {notebook_name} no encontrado en {NOTEBOOKS_DIR}")
        return False

    try:
        print(f"  ‚úÖ Ejecutando {notebook_name}...")
        %run $path

        if not validar_salidas(notebook_name):
            print(f"    ‚ùå {notebook_name} no gener√≥ los archivos esperados.")
            return False

        return True
    except Exception as e:
        print(f"    ‚ùå Error: {str(e)[:200]}...")
        return False

# --- FUNCI√ìN PARA GENERAR PDF MAESTRO ---
def generate_master_report():
    """Genera un PDF maestro con toda la informaci√≥n del sistema de agentes."""
    print("üìÑ Generando PDF maestro con toda la informaci√≥n...")

    try:
        from markdown import markdown
        from weasyprint import HTML

        # --- Leer todos los datos generados ---
        master_data = {}

        # Datos de cartera
        portfolio_path = f"{REPORTS_DIR}/portfolio_enriched_final.csv"
        if os.path.exists(portfolio_path):
            portfolio_df = pd.read_csv(portfolio_path)
            master_data['portfolio_valor'] = portfolio_df['importe_actual_eur'].sum()
            master_data['portfolio_composicion'] = portfolio_df[['nombre', 'importe_actual_eur', 'peso_%']].to_dict('records')

        # Risk dashboard
        risk_path = f"{REPORTS_DIR}/risk_dashboard.csv"
        if os.path.exists(risk_path):
            risk_df = pd.read_csv(risk_path)
            master_data['risk'] = {}
            for _, row in risk_df.iterrows():
                if 'm√©trica' in row and 'valor' in row:
                    master_data['risk'][row['m√©trica']] = row['valor']
                elif 'metrica' in row and 'valor' in row:  # Sin acento
                    master_data['risk'][row['metrica']] = row['valor']

        # Market regime
        market_path = f"{REPORTS_DIR}/market_regime_latest.json"
        if os.path.exists(market_path):
            with open(market_path, 'r') as f:
                master_data['market'] = json.load(f)

        # Liquidity regime
        liquidity_path = f"{REPORTS_DIR}/liquidity_regime_latest.json"
        if os.path.exists(liquidity_path):
            with open(liquidity_path, 'r') as f:
                master_data['liquidity'] = json.load(f)

        # Sectorial strength
        sector_path = f"{REPORTS_DIR}/sector_strength_latest.json"
        if os.path.exists(sector_path):
            with open(sector_path, 'r') as f:
                master_data['sector'] = json.load(f)

        # FX signals
        fx_path = f"{REPORTS_DIR}/fx_hedge_latest.json"
        if os.path.exists(fx_path):
            with open(fx_path, 'r') as f:
                master_data['fx'] = json.load(f)

        # Quant signals
        quant_path = f"{REPORTS_DIR}/quant_signals_latest.json"
        if os.path.exists(quant_path):
            with open(quant_path, 'r') as f:
                master_data['quant'] = json.load(f)

        # --- Construir HTML maestro ---
        html_content = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <meta charset="utf-8">
            <title>Informe Maestro de Inversi√≥n - {datetime.today().strftime('%Y-%m-%d')}</title>
            <style>
                body {{ font-family: Arial, sans-serif; margin: 40px; line-height: 1.6; }}
                h1, h2, h3 {{ color: #2c3e50; }}
                .section {{ margin-bottom: 30px; padding: 20px; border-left: 4px solid #3498db; background-color: #f8f9fa; }}
                table {{ border-collapse: collapse; width: 100%; margin: 20px 0; }}
                th, td {{ border: 1px solid #bdc3c7; padding: 12px; text-align: left; }}
                th {{ background-color: #ecf0f1; }}
                .emoji {{ font-size: 1.2em; }}
                .highlight {{ background-color: #fff3cd; padding: 15px; border-radius: 5px; margin: 10px 0; }}
            </style>
        </head>
        <body>
            <h1>üìä Informe Maestro de Inversi√≥n</h1>
            <p><strong>Fecha:</strong> {datetime.today().strftime('%Y-%m-%d')}</p>

            <!-- RESUMEN EJECUTIVO -->
            <div class="section">
                <h2>üéØ Resumen Ejecutivo</h2>
        """

        # Resumen ejecutivo
        valor_total = master_data.get('portfolio_valor', 0)
        drawdown = master_data['risk'].get('Drawdown actual', 'N/A') if 'risk' in master_data else 'N/A'
        alpha = master_data['risk'].get('Alpha vs S&P 500', 'N/A') if 'risk' in master_data else 'N/A'
        regimen_mercado = master_data['market'].get('regimen', 'N/A') if 'market' in master_data else 'N/A'
        regimen_liquidez = master_data['liquidity'].get('regimen', 'N/A') if 'liquidity' in master_data else 'N/A'

        html_content += f"""
                <p><strong>Valor total de cartera:</strong> {valor_total:,.0f} ‚Ç¨</p>
                <p><strong>Drawdown actual:</strong> {drawdown}</p>
                <p><strong>Alpha vs S&P 500:</strong> {alpha}</p>
                <p><strong>R√©gimen de mercado:</strong> {regimen_mercado}</p>
                <p><strong>R√©gimen de liquidez:</strong> {regimen_liquidez}</p>
            </div>

            <!-- AN√ÅLISIS MACRO -->
            <div class="section">
                <h2>üåê An√°lisis Macro</h2>
        """

        if 'market' in master_data:
            html_content += f"""
                <p><strong>R√©gimen de Mercado:</strong> {master_data['market']['regimen']} (Score: {master_data['market']['score']})</p>
                <p><strong>Percentil Hist√≥rico:</strong> {master_data['market'].get('percentil', 'N/A')}%</p>
            """

        if 'liquidity' in master_data:
            html_content += f"""
                <p><strong>R√©gimen de Liquidez:</strong> {master_data['liquidity']['regimen']} (Score: {master_data['liquidity']['score']}/7)</p>
                <h3>Riesgos Cr√≠ticos:</h3>
                <ul>
        """
            riesgos = master_data['liquidity'].get('riesgos', [])
            riesgos_criticos = [r for r in riesgos if r.get('alerta', False)]
            if riesgos_criticos:
                for riesgo in riesgos_criticos:
                    html_content += f"<li>{riesgo['nombre']}: {riesgo['valor']} ({riesgo['nivel']})</li>"
            else:
                html_content += "<li>No hay riesgos cr√≠ticos identificados.</li>"
            html_content += "</ul>"

        html_content += """
            </div>

            <!-- SE√ëALES DE INVERSI√ìN -->
            <div class="section">
                <h2>üìà Se√±ales de Inversi√≥n</h2>
        """

        # Se√±ales sectoriales
        if 'sector' in master_data and master_data['sector'].get('top_oportunidades'):
            html_content += "<h3>Top Oportunidades Sectoriales:</h3><ul>"
            for sector in master_data['sector']['top_oportunidades']:
                html_content += f"<li>{sector}</li>"
            html_content += "</ul>"

        # Se√±ales FX
        if 'fx' in master_data and master_data['fx'].get('divisas_a_cubrir'):
            html_content += "<h3>Se√±ales FX:</h3><ul>"
            for divisa in master_data['fx']['divisas_a_cubrir']:
                recomendacion = master_data['fx']['recomendaciones'].get(divisa, {})
                cobertura = recomendacion.get('%_cobertura', 'N/A')
                html_content += f"<li>{divisa}: {cobertura}% cobertura recomendada</li>"
            html_content += "</ul>"

        # Se√±ales cuantitativas
        if 'quant' in master_data and master_data['quant'].get('se√±ales'):
            html_content += "<h3>Se√±ales Cuantitativas:</h3><ul>"
            for activo, se√±al in master_data['quant']['se√±ales'].items():
                html_content += f"<li>{activo}: {se√±al}</li>"
            html_content += "</ul>"

        html_content += """
            </div>

            <!-- COMPOSICI√ìN DE CARTERA -->
            <div class="section">
                <h2>üíº Composici√≥n de Cartera</h2>
        """

        if 'portfolio_composicion' in master_data:
            html_content += """
                <table>
                    <tr><th>Producto</th><th>Valor (‚Ç¨)</th><th>% Cartera</th></tr>
            """
            for item in master_data['portfolio_composicion']:
                html_content += f"<tr><td>{item['nombre']}</td><td>{item['importe_actual_eur']:,.0f}</td><td>{item['peso_%']:.1f}%</td></tr>"
            html_content += "</table>"

        html_content += """
            </div>

            <p><em>Informe generado autom√°ticamente por el Sistema de Agentes de Inversi√≥n.</em></p>
        </body>
        </html>
        """

        # Guardar PDF maestro
        pdf_path = f"{REPORTS_DIR}/master_report_{datetime.today().strftime('%Y%m%d')}.pdf"
        HTML(string=html_content).write_pdf(pdf_path)
        print(f"‚úÖ PDF maestro guardado en: {pdf_path}")
        return True

    except Exception as e:
        print(f"‚ö†Ô∏è Error al generar PDF maestro: {e}")
        return False

def generate_ai_dashboard_summary():
    """
    Genera un an√°lisis exhaustivo de la cartera usando GenAI,
    con sem√°foro, recomendaci√≥n y justificaci√≥n por indicador.
    """
    print("üß† Generando an√°lisis exhaustivo con GenAI...")

    # --- Cargar datos con manejo robusto ---
    pf_path = f"{REPORTS_DIR}/portfolio_enriched_final.csv"
    pf = pd.read_csv(pf_path) if os.path.exists(pf_path) else pd.DataFrame()

    risk_path = f"{REPORTS_DIR}/risk_dashboard.csv"
    risk = pd.read_csv(risk_path) if os.path.exists(risk_path) else pd.DataFrame()

    quant_path = f"{REPORTS_DIR}/quant_signals.csv"
    quant = pd.read_csv(quant_path) if os.path.exists(quant_path) else pd.DataFrame()

    fx_path = f"{REPORTS_DIR}/fx_hedge_signal.csv"
    fx = pd.read_csv(fx_path) if os.path.exists(fx_path) else pd.DataFrame()

    metrics_path = f"{REPORTS_DIR}/asset_metrics.csv"
    metrics = pd.read_csv(metrics_path) if os.path.exists(metrics_path) else pd.DataFrame()

    sector_path = f"{REPORTS_DIR}/sector_scores_latest.csv"
    sectors = pd.read_csv(sector_path) if os.path.exists(sector_path) else pd.DataFrame()

    breaches_asset_path = f"{REPORTS_DIR}/breaches_asset.csv"
    breaches_asset = pd.DataFrame()
    if os.path.exists(breaches_asset_path):
        try:
            breaches_asset = pd.read_csv(breaches_asset_path, encoding='utf-8-sig')
            breaches_asset.columns = breaches_asset.columns.str.strip()
        except Exception as e:
            print(f"‚ö†Ô∏è Error al leer breaches_asset.csv: {e}")

    # --- Funci√≥n robusta para extraer m√©tricas del risk dashboard ---
    def get_risk_metric(risk_df, metric_name):
        """Extrae una m√©trica del risk dashboard de forma robusta."""
        if risk_df.empty:
            return "N/A"

        # Encontrar la columna de m√©trica (maneja diferentes nombres)
        metric_col = None
        value_col = None

        for col in risk_df.columns:
            col_lower = col.lower().strip()
            if 'm√©trica' in col_lower or 'metrica' in col_lower or 'metric' in col_lower:
                metric_col = col
            elif 'valor' in col_lower or 'value' in col_lower:
                value_col = col

        if not metric_col or not value_col:
            return "N/A"

        # Buscar la fila que contiene la m√©trica
        mask = risk_df[metric_col].astype(str).str.contains(metric_name, case=False, na=False)
        if mask.any():
            return risk_df.loc[mask, value_col].iloc[0]
        return "N/A"

    # --- Extraer m√©tricas clave ---
    total = pf["importe_actual_eur"].sum() if not pf.empty else 0
    drawdown = get_risk_metric(risk, 'Drawdown actual')
    alpha = get_risk_metric(risk, 'Alpha vs S&P 500')
    retorno_cartera = get_risk_metric(risk, 'Retorno anualizado cartera')
    sp500_retorno = get_risk_metric(risk, 'Retorno anualizado S&P 500')

    activos_concentrados = "N/A"
    regiones_concentradas = "N/A"
    if not breaches_asset.empty:
        # Encontrar columnas correctas en breaches_asset
        metric_col = None
        value_col = None
        for col in breaches_asset.columns:
            col_lower = col.lower().strip()
            if 'm√©trica' in col_lower or 'metrica' in col_lower or 'metric' in col_lower:
                metric_col = col
            elif 'valor' in col_lower or 'value' in col_lower:
                value_col = col

        if metric_col and value_col:
            activos_row = breaches_asset[breaches_asset[metric_col].astype(str).str.contains('Activos > 5%', case=False, na=False)]
            regiones_row = breaches_asset[breaches_asset[metric_col].astype(str).str.contains('Regiones > 30%', case=False, na=False)]
            if not activos_row.empty:
                activos_concentrados = activos_row[value_col].iloc[0]
            if not regiones_row.empty:
                regiones_concentradas = regiones_row[value_col].iloc[0]

    usd_no_hedged = "N/A"
    if not breaches_asset.empty and metric_col and value_col:
        usd_row = breaches_asset[breaches_asset[metric_col].astype(str).str.contains('USD no hedged', case=False, na=False)]
        if not usd_row.empty:
            usd_no_hedged = usd_row[value_col].iloc[0]

    # --- Construir prompt para GenAI ---
    prompt = f"""
Eres un agente de inversi√≥n profesional con acceso a un sistema multiagente avanzado.
Tu tarea es generar un **an√°lisis exhaustivo de la cartera al {datetime.today().strftime('%Y-%m-%d')}** en formato de tabla con las siguientes columnas:

- **Categor√≠a**: riesgo, rendimiento, concentraci√≥n, FX, sectorial, se√±ales, etc.
- **Indicador**: nombre del indicador clave.
- **Valor**: valor num√©rico o categ√≥rico.
- **Sem√°foro**: usa ‚úÖ (verde), ‚ö†Ô∏è (√°mbar) o ‚ùå (rojo).
- **¬øVamos bien?**: "S√≠", "No" o "Parcialmente".
- **Recomendaci√≥n**: acci√≥n concreta (ej. "Rebalancear", "Aumentar XLK", "Cubrir 10% USD").
- **Justificaci√≥n**: explica por qu√©, citando el indicador y su valor.

**Reglas estrictas**:
- Usa **solo los datos proporcionados**.
- S√© **conciso, preciso y accionable**.
- No inventes datos.
- El tono debe ser profesional pero claro para un inversor no t√©cnico.

**Datos de la cartera**:
- Valor total: {total:,.0f} ‚Ç¨
- Drawdown actual: {drawdown}
- Retorno anualizado cartera: {retorno_cartera}
- Retorno anualizado S&P 500: {sp500_retorno}
- Alpha vs S&P 500: {alpha}
- Activos >5% del portafolio: {activos_concentrados}
- Regiones >30%: {regiones_concentradas}
- USD no hedged: {usd_no_hedged}%

**Se√±ales sectoriales (Symbol, Score, RSC, Signal)**:
{sectors.to_string(index=False) if not sectors.empty else "No disponible"}

**Se√±ales cuantitativas (ejemplos)**:
{quant.head(5).to_string(index=False) if not quant.empty else "No disponible"}

**Brechas de riesgo**:
- Drawdown ‚â•9%: {'S√≠' if drawdown != 'N/A' and 'True' in str(drawdown) else 'No'}
- Alpha objetivo (+5%): {'Cumplido' if alpha != 'N/A' and isinstance(alpha, str) and float(str(alpha).replace('%','').replace(',','.')) >= 5 else 'Incumplido'}

Genera la tabla en **formato Markdown**. No a√±adas introducci√≥n ni conclusi√≥n, solo la tabla.
"""

    try:
        response = model.generate_content(prompt)
        analysis_md = response.text

        output_path = f"{REPORTS_DIR}/ai_dashboard_summary_{datetime.today().strftime('%Y%m%d')}.md"
        with open(output_path, "w", encoding="utf-8") as f:
            f.write("# üìä An√°lisis Exhaustivo de Cartera (GenAI)\n\n")
            f.write(f"Fecha: {datetime.today().strftime('%Y-%m-%d')}\n\n")
            f.write(analysis_md)

        print(f"‚úÖ An√°lisis guardado en: {output_path}")
        return analysis_md

    except Exception as e:
        error_msg = f"‚ö†Ô∏è Error al generar an√°lisis con GenAI: {e}"
        print(error_msg)
        return error_msg

# --- Funci√≥n para ejecutar la cadena semanal ---
def run_weekly_pipeline():
    print("üîÑ Iniciando cadena semanal de agentes...")
    log = []
    start_time = datetime.now()

    for nb in NOTEBOOKS:
        success = run_notebook(nb)
        log.append(f"{nb}: {'OK' if success else 'FALL√ì'}")
        if not success:
            print("  ‚ö†Ô∏è Cadena interrumpida por error.")
            break

    end_time = datetime.now()
    duration = (end_time - start_time).total_seconds()
    log_path = f"{REPORTS_DIR}/orchestrator_log_{start_time.strftime('%Y%m%d')}.txt"
    with open(log_path, "w") as f:
        f.write(f"Cadena ejecutada: {start_time}\n")
        f.write(f"Duraci√≥n: {duration:.0f} segundos\n")
        f.write("\n".join(log))

    print(f"‚úÖ Cadena completada en {duration:.0f} segundos. Log guardado en {log_path}")

    if all("OK" in entry for entry in log):
        print("üß† Generando an√°lisis exhaustivo con GenAI...")
        generate_ai_dashboard_summary()

        print("üìÑ Generando PDF maestro con toda la informaci√≥n...")
        generate_master_report()
    else:
        print("‚ö†Ô∏è No se gener√≥ el an√°lisis de GenAI ni el PDF maestro debido a errores en la cadena.")

# --- Funci√≥n para responder preguntas ---
def answer_question(user_question):
    try:
        # --- Cargar datos existentes con manejo de errores ---
        pf_path = f"{REPORTS_DIR}/portfolio_enriched_final.csv"
        pf = pd.read_csv(pf_path) if os.path.exists(pf_path) else pd.DataFrame()

        risk_path = f"{REPORTS_DIR}/risk_dashboard.csv"
        risk = pd.read_csv(risk_path) if os.path.exists(risk_path) else pd.DataFrame()

        # --- Depuraci√≥n: mostrar columnas reales ---
        if not risk.empty:
            print(f"üîç Columnas reales en risk_dashboard: {list(risk.columns)}")

        quant_path = f"{REPORTS_DIR}/quant_signals.csv"
        quant = pd.read_csv(quant_path) if os.path.exists(quant_path) else pd.DataFrame()

        fx_path = f"{REPORTS_DIR}/fx_hedge_signal.csv"
        fx = pd.read_csv(fx_path) if os.path.exists(fx_path) else pd.DataFrame()

        metrics_path = f"{REPORTS_DIR}/asset_metrics.csv"
        metrics = pd.read_csv(metrics_path) if os.path.exists(metrics_path) else pd.DataFrame()

        # --- Encontrar la columna correcta para m√©trica ---
        metric_col = None
        value_col = None

        if not risk.empty:
            # Buscar columna que contenga 'm√©trica', 'metrica', o similar
            for col in risk.columns:
                if 'm√©trica' in col.lower() or 'metrica' in col.lower() or 'metric' in col.lower():
                    metric_col = col
                    break

            # Buscar columna que contenga 'valor'
            for col in risk.columns:
                if 'valor' in col.lower() or 'value' in col.lower():
                    value_col = col
                    break

        # --- Cargar contexto MACRO ---
        macro_str = "No disponible"
        macro_path = f"{REPORTS_DIR}/market_regime_latest.json"
        if os.path.exists(macro_path):
            with open(macro_path) as f:
                macro = json.load(f)
            macro_str = f"{macro['regimen']} (Score: {macro['score']:.2f}, Percentil: {macro['percentil']:.0f}%)"

        # --- Cargar fuerza SECTORIAL ---
        sectors_str = "No disponible"
        sectors_path = f"{REPORTS_DIR}/sector_strength_latest.json"
        if os.path.exists(sectors_path):
            with open(sectors_path) as f:
                sectors_data = json.load(f)
            if sectors_data.get("top_oportunidades"):
                sectors_str = "; ".join(sectors_data["top_oportunidades"])

        # --- Extraer KPIs de riesgo ---
        drawdown = "N/A"
        alpha = "N/A"

        if not risk.empty and metric_col and value_col:
            # Buscar Drawdown actual
            dd_rows = risk[risk[metric_col].astype(str).str.contains('Drawdown', case=False, na=False)]
            if not dd_rows.empty:
                drawdown = dd_rows[value_col].iloc[0]

            # Buscar Alpha
            alpha_rows = risk[risk[metric_col].astype(str).str.contains('Alpha', case=False, na=False)]
            if not alpha_rows.empty:
                alpha = alpha_rows[value_col].iloc[0]

        # --- Resto del c√≥digo igual ---
        total = pf["importe_actual_eur"].sum() if not pf.empty else 0
        quant_summary = "Ninguna" if quant.empty else ", ".join([f"{row['Activo']}: {row['Se√±al']}" for _, row in quant.iterrows()])
        fx_summary = "Ninguna" if fx.empty else ", ".join([f"{row['divisa']}: {row['cobertura_recomendada']} ({row['%_cobertura']}%)" for _, row in fx.iterrows()])
        acn_alpha = "N/A"
        if not metrics.empty and "ACN" in metrics["Activo"].values:
            acn_row = metrics[metrics["Activo"] == "ACN"].iloc[0]
            acn_alpha = f"{acn_row['Retorno total mejorado']:.1%}" if pd.notna(acn_row.get('Retorno total mejorado')) else "N/A"

        context = f"""
        Eres un asistente de inversi√≥n personal. Responde usando SOLO estos datos:

        **Contexto Macro Actual:**
        - R√©gimen: {macro_str}

        **Fuerza Sectorial (Top oportunidades):**
        - {sectors_str}

        **Cartera Global:**
        - Valor total: {total:,.0f} ‚Ç¨
        - Drawdown: {drawdown}
        - Alpha vs S&P 500: {alpha}

        **Se√±ales Cuantitativas:**
        - {quant_summary}

        **Se√±ales FX:**
        - {fx_summary}

        **Insights por Activo:**
        - Alpha de ACN (incluyendo regalos): {acn_alpha}

        Pregunta del usuario: {user_question}
        """

        response = model.generate_content(context)
        return response.text

    except Exception as e:
        return f"‚ö†Ô∏è Error al cargar datos: {e}"

# --- Interfaz principal ---
print("üß† Orchestrator + Agente GenAI Activo")
print("Comandos:")
print("- 'semanal': ejecuta la cadena completa de agentes")
print("- 'pregunta [tu pregunta]': interact√∫a con GenAI")
print("- 'salir': termina")

while True:
    user_input = input("\n> ").strip()

    if user_input == "salir":
        break
    elif user_input == "semanal":
        run_weekly_pipeline()
    elif user_input.startswith("pregunta "):
        question = user_input[9:]
        answer = answer_question(question)
        print(f"\nü§ñ {answer}")
    else:
        print("‚ùì Comando no reconocido. Usa 'semanal', 'pregunta ...' o 'salir'.")
```
